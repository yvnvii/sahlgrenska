```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
import tensorflow as tf

# Start from your original diq_df (replace this line if already loaded)
# diq_df = pd.read_sas("DIQ_J.xpt")

# Filter and map target
filtered_df = diq_df[diq_df["Told_Have_Diabetes"].isin([1, 2])].copy()
filtered_df["Diabetes_Binary"] = filtered_df["Told_Have_Diabetes"].map({1: 1, 2: 0})

# Balance the data
df_majority = filtered_df[filtered_df.Diabetes_Binary == 0]
df_minority = filtered_df[filtered_df.Diabetes_Binary == 1]
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority])

# Prepare features
X = df_balanced.drop(columns=["Respondent_ID", "Told_Have_Diabetes", "Diabetes_Binary"])
X = X.select_dtypes(include=[np.number]).dropna(axis=1, how='all')
X = X.fillna(X.median())
y = df_balanced["Diabetes_Binary"]

# Normalize & split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build neural network
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(X_train, y_train, validation_split=0.2, epochs=20, batch_size=32)

# Evaluate
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```
